## Apply, applymap, map

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3, 4],
                   'B': [10, 20, 30, 40],
                   'C': [20, 40, 60, 80]
                   },
                  index=['Linha 1', 'Linha 2', 'Linha 3', 'Linha 4'])
```

```python
df
```

O `apply`  usado para aplicar uma funo ao longo de um eixo do Dataframe ou em valores de Srie

```python
def custom_sum(row):
    return row.sum()

#abaixo; Cria uma coluna 'D' com a somatria na horizontal(Linhas)
df['D'] = df.apply(custom_sum, axis=1) # ou df['D'] = df.apply(lambda row: row.sum(), axis=1) # apply em um dataframe
```

```python
df
```

```python
df.loc['Linha 5'] = df.apply(custom_sum, axis=0)   #cria uma linha  chamada 'linha5'. Soma nas colunas(verticais.)
df
```

```python
def multiply_by_2(val):
    return val * 2


df['E'] = df['C'].apply(multiply_by_2) # apply sobre uma srie df['E'] = df['C'].apply(lambda x: x * 2)
```

```python
df
```

O `applymap` s est disponvel no Dataframe e  usado para operaes de elemento a elemento em todo o DataFrame

```python
df.applymap(np.square)
```

O `map` s est disponvel e, sries e  usado para substituir cada valor em uma srie por outro valor

```python
s = pd.Series (['gato', 'cachorro', np.nan, 'coelho']) 
s
```

```python
s.map({'gato': 'gatinho', 'cachorro':'cachorrinho', 'coelho':'coelhinho'}).fillna('aranha')
```

```python
s.map('Eu sou um {}'.format, na_action='ignore')
```

## Dataframe com linhas do tipo objeto (string)

```python
df = pd.DataFrame({'email': ['jose@email.com', 'joao@email.com', 'maria@email.com', 'joana@email.com'],
                   'nome': ['Jose', 'Joao', 'Maria', 'Joana'],
                   'idade': [20, 40, 60, 30],
                   'telefone': ['99999-1111', '99999-2222','99999-3333','99999-4444']
                   })
```

```python
df
```

```python
df.info
```

```python
#criando uma coluna chamada 'nome_upper' transformando as strings em letras maisculas(.str.upper)
df['nome_upper'] = df['nome'].str.upper()  #quero tudo em letras maisculas.  O '.str' acessas as strings da coluna(pode ser feito no dataframe inteiro se interessar)
```

```python
df
```

```python
df['nome'].str.lower()  #visualizao em letras minsculas.
```

```python
#criando columa com a quantidade de caracteres da coluna nome.
df['qt_caracters_nome'] = df['nome'].str.len()
```

```python
df
```

```python
# trocar maria por mariana
df['nome_upper'] = df['nome_upper'].str.replace('MARIA','MARIANA')

# o replace  para o dataframe. O str  para as strings na coluna especificada.
```

```python
df
```

```python
df['telefone'].str.strip() # elemina os espaos em excesso
```

```python
df['nome'].str[0:2] # acessar os caracteres das strings da coluna nome
```

```python
# tentando transformar a coluna em numrica
pd.to_numeric(df['telefone'])

#d erro por causa do trao(-) do nmero. O comando pd.to_numeric no consegue transformar "-" em nmero.
```

```python
# exemplo1  o split vai separar a string a partir do caracter indicado, nesse caso o '-'
df['telefone'].str.split('-')
```

```python
# exemplo1  Pode-se utilizar o argumento expand=True para separar em duas colunas.
df['telefone'].str.split('-', expand=True)
```

```python
# exemplo1 - Pode-se utilizar o comando get para pegar uma das partes da string:
df['telefone'].str.split('-').str.get(0)
```

```python
# exemplo1 Pode-se utilizar o comando get para pegar uma das partes da string:
df['telefone'].str.split('-').str.get(1)
```

```python
#exemplo2  Para transformar o '-' em vazio ''. Desta maneira o '-' desaparece e o telefone  transfomado em nmero(ainda do tipo objeto)
df['telefone_ajust'] = df['telefone'].str.replace('-','')
```

```python
df['telefone_ajust'] 
```

```python
#Agora vamos transformar o telefone, coluna telefone_ajust, em int64.
pd.to_numeric(df['telefone_ajust'])         # ou df['telefone_ajust'].astype('int') 
```

```python
df
```

```python
df[['nomedoemail','dominiodoemail']] =df['email'].str.split('@', expand=True)

# df[['nomedoemail','dominiodoemail']]  cria duas colunas, 'nomedoemail' e 'dominiodoemail'
# '@'  a string ponto de separao
# expand=True -> para criao de duas colunas

```

## TITANIC - DATASET

https://www.kaggle.com/c/titanic/data

| Variable | Definition                                 | Key                                            |
|----------|--------------------------------------------|------------------------------------------------|
| survival | Survival                                   | 0 = No, 1 = Yes                                |
| pclass   | Ticket class                               | 1 = 1st, 2 = 2nd, 3 = 3rd                      |
| sex      | Sex                                        |                                                |
| Age      | Age in years                               |                                                |
| sibsp    | # of siblings / spouses aboard the Titanic |                                                |
| parch    | # of parents / children aboard the Titanic |                                                |
| ticket   | Ticket number                              |                                                |
| fare     | Passenger fare                             |                                                |
| cabin    | Cabin number                               |                                                |
| embarked | Port of Embarkation                        | C = Cherbourg, Q = Queenstown, S = Southampton |

**Variable Notes**  
pclass: A proxy for socio-economic status (SES)  
1st = Upper  
2nd = Middle  
3rd = Lower  

age: Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5

sibsp: The dataset defines family relations in this way...  
Sibling = brother, sister, stepbrother, stepsister  
Spouse = husband, wife (mistresses and fiancs were ignored)  

parch: The dataset defines family relations in this way...  
Parent = mother, father  
Child = daughter, son, stepdaughter, stepson  
Some children travelled only with a nanny, therefore parch=0 for them

```python
# dummies # pandas.get_dummies(data, prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None)
#transformando  C Q S em colunas para termos zeros e uns para podermos contar.
#com isso conseguimos transformar uma coluna categrica em 3 colunas numricas.

pd.get_dummies(df['Embarked'])
```

```python
# eliminar a coluna Embarked
df = df.drop(columns=['Embarked'])
df
```

```python
# trocar o nome da coluna C
df.rename(columns={'C':'Embarked_C'}, inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True) # preenchendo as idades faltantes com a mediana das idades

df['Age'] = df['Age'].astype(int) # transformando as idades em inteiro
```

```python
# criar 5 faixas de idade

df['faixas_idade'] = pd.cut(df['Age'], 5) # ver tb o #pd.qcut
```

```python
df.groupby('faixas_idade')['Survived'].sum()
```

```python
# a mdia de sobrevivncia de classe e sexo em uma tabela

df.groupby(['Sex','Pclass'])[['Survived']].mean().unstack()
```
