## Missing data - Valores faltantes

Inspecionar os dados faltantes  
NaN --> Not a number - não um número  
NaT --> Not a timestamp - não é formato data  

```python
# contando os nulos, somando os valores nulos NaN
df.isnull().sum() # ou df.isna().sum()
```

```python
df.isnull() # mostra o dataframe inteiro e retorna true ou false para a condição nulo.
```

```python
df.isnull().count().sort_values(ascending=False)  ##Conta o dataframe inteiro(vide acima), não apenas que me é nulo.
```

```python
(df.isnull().sum()/df.isnull().count()).sort_values(ascending=False) #divide o total nulo, pelo total do dataframe.
```

```python
# verificar missing data
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total','Percent'])
missing_data
```

```python
# Verificando os dados faltantes por exemplo da coluna open no dataframe
df.loc[df['Open'].isnull()] 
```

```python
# outra forma de ver os dados faltantes
# !pip install missingno
import missingno as msno
msno.matrix(df)
```

Tratar os dados faltantes - limpeza dos dados

- Atribuir um valor a eles - preencher com a média, o valor anterior, regressão linear, colocar um valor aleatório
- Eliminar os valores faltantes - ATENÇÃO o valor pertence a uma linha ou uma coluna... excluir a linha ou coluna

```python
# Eliminando os dados

df.dropna(how='any') # eliminando todas as linhas que tem dados faltantes
```

```python
# preencher os dados
df.fillna(0) # preencher dados faltantes com valor valor zero
```

```python
df['Open'].fillna(df['Open'].mean()) # preencher dados faltantes com média, mediana, regressão linear
```

```python
# ffill: propagate last valid observation forward to next valid backfill / 
# bfill: use next valid observation to fill gap.

df.fillna(method='bfill') # preenche os dados com o dado anterior
```

```python
df = df.fillna(method='ffill') # preenche os dados com o dado anteriores
```

```python
msno.matrix(df)
```

```python
df = df.dropna(how='any')
```

```python
msno.matrix(df)
```

## Apply, applymap e map

- apply() é usado para aplicar uma função ao longo do eixo do **DataFrame** ou em valores de **Série**
- applymap() é usado para aplicar uma função a um **Dataframe** elemento a elemento
- map() é usado para substituir cada valor em uma **Série** por outro valor

```python
# Vou escrever 3 formas diferentes de montar uma função (amplitude)
    
def funcao(x): # x é um array
    resultado = x.max() - x.min()
    return resultado

def funcao(x):
    return x.max() - x.min()

funcao: lambda x: x.max() - x.min()
```

```python
# apply - quero aplicar a função em cima desse dataframe - que no caso quero saber a amplitude dos dados
df[['Close','Low','High','Open']].apply(funcao) # ou df[['Close','Low','High','Open']].apply(lambda x: x.max() - x.min())
```

```python
# Falta exemplo do apply para serie
```

```python
f1 = lambda x: '%.2f' % x     #arredonda os valores de x em duas casas decimais

df[['Open','Close']].applymap(f1)
```

```python
# outro exemplo do applymap
df[['Open','Close']].applymap(np.sqrt) #quero tirar a raiz quadrada de cada um dos elementos
```

```python
df['Open'].map(np.sqrt)
```
