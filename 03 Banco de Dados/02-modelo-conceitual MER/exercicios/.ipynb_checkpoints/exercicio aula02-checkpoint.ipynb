## Enunciado

Uma escola de idiomas deseja automatizar seus processos.

Essa escola oferece cursos, sendo que cada curso pode ser ministrado por apenas um professor.

Os alunos podem se inscrever nos cursos e participar de mais de um curso ao mesmo tempo.

A escola possui informações dos alunos, como RG, Nome, Endereço, Data de Nascimento, Idade, Gênero e Data da matrícula dos alunos.

Os seguintes dados dos professores também devem ser armazenados: RG, Nome, Endereço, Data de Nascimento, Idade, Gênero, Disponibilidade, Habilidades e Data de admissão.

Além disso, o sistema deve gerenciar o nome e a descrição do curso.

Q1
Quais são as entidades, atributos e os relacionamentos que podemos inferir desse problema?

Q2
Criar um diagrama ER que represente a modelagem conceitual deste problema.

NOTAS DE RESOLUÇÃO:

Considerei que curso é diferente de turma. Uma escola pode ter os seguintes cursos(exemplo): inglês, espanhol e francês.  Cada um dos curso pode ter N turmas.
Considerei que a escola pode ter cursos a distância(Não presenciais), cursos ao vivo, cursos gravados.  Considerando que há cursos gravados, um aluno não precisa necessariamente pertencer a uma turma.
A escola também tem os cursos tradicionais(presenciais).

Para o requisito do exercicio que diz que cada curso pode ser ministrado por apenas um professor, a solução endereçará isso através da entidade turma, ou seja, cada professor terá apenas uma turma.

Para o requisito do exercício que exige que  o sistema deve gerenciar o nome, teremos o atributo "disciplina" que fará a vez do do atributo "nome".


## Q1 Quais são as entidades, atributos e os relacionamentos que podemos inferir desse problema?

Entidade:curso
Atributos da entidade curso: duração,descrição,investimento,nível(estágio),código(chave), disciplina.
Entidades filhas:remoto, presencial, gravado,ao-vivo.

Entidade:professor
Atributos da entidade professor:ID_PROF(chave), RG, CPF, Endereço(endereço, número, complemento, CEP, cidade, estado), Data de Nascimento, Idade, Gênero, Disponibilidade, salário, Nome(1o Nome, 2o Nome, 3o Nome), habilidades.

Entidade: aluno
Atributos da enteidade aluno:Matrícula(chave), CPF, RG, Nome(1o Nome, 2o Nome, 3o Nome),Endereço(endereço, número, complemento, CEP, cidade, estado), Data de Nascimento, Idade, Gênero, Data de Matrícula.

Entidade:turma
Atributos da entidade Turma: horário, dias de aula,turma(chave), ID Prof(chave estrangeira)
