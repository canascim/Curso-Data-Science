HTTP - HyperText Transfer Protocol

```
                  request
       ---------------------------------->
cliente                                  servidor
(client)                                 (server)
       <----------------------------------
                  response

GET
POST
PUT

status code
2xx - "sucesso": 200
3xx - "redirecionamento"
4xx - "página não encontrada": 404, "forbidden": 403
5xx - erros do servidor
```

API

````python
# Nossa primeira requisição!

import requests

resposta = requests.get('https://letscode.com.br/')

# o atributo .text do objeto "resposta" contém a "payload" da resposta do servidor interpretada como texto!

print('Status code:', resposta.status_code, '\n\n\n')

print(resposta.text)
````

````python
# e o tal do status_code?

usuario = input('Digite uma subpágina:')

resposta = requests.get('https://letscode.com.br/' + usuario)

if resposta.status_code == 200:
    print(resposta.text)
else:
    print('Atenção! Erro:', resposta.status_code)
````

````python
# Numbers API

import requests

# vamos pedir um fato matemático sobre o número 10
resposta = requests.get('http://numbersapi.com/10/math')

if resposta.status_code == 200:
    print(resposta.text)
else:
    print('Atenção! Erro:', resposta.status_code)
````

```

    'nome':'José',
```

````python
import requests

resposta = requests.get('https://swapi.dev/api/people/4/')

if resposta.status_code == 200:
    dicionario = resposta.json()
    print(dicionario['name'])
else:
    print('Erro! Código:', resposta.status_code)
````

````python
# request a partir de outro request

resposta = requests.get('https://swapi.dev/api/people/4/')

if resposta.status_code == 200:
    dicionario = resposta.json()
    respostaPlaneta = requests.get(dicionario['homeworld'])
    if respostaPlaneta.status_code == 200:
        dicionarioPlaneta = respostaPlaneta.json()
        print(dicionario['name'], '-', dicionarioPlaneta['name'])
else:
    print('Erro! Código:', resposta.status_code)
````
