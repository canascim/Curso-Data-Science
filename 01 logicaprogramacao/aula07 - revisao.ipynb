```python
# essa função não foi preparada para receber listas!
def somaDois(x):
    return 2 + x

# funciona pra int!
maisdois = somaDois(10)
print(maisdois)

# não funciona pra lista...
lista = [1, 2, 3, 4]
dlista = somaDois(lista)
print(dlista)
    
```

```python
# essa função SÓ funciona com listas e similares
def somatorio(x):
    soma = 0
    for numero in x:
        soma = soma + numero
        
    return soma

lista = [1, 2, 3, 4]
s = somatorio(lista)
print(s)

s2 = somatorio(10)
print(s2)
```
