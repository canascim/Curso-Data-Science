```python
# listas!

lista = [10, 3.14, 'Rafael', True, 7, 2, 10]
#índice= 0     1      2        3   4  5   6
print(lista)

# índices: começam no zero e crescem de um em um
print(lista[4])
print(lista[1])

lista[1] = 'três ponto quatorze'
print(lista)

# podemos adicionar e remover elementos da lista:
lista.append('valor novo')
lista.append('novíssimo')
lista.append(5)
lista.remove(10)
print(lista)

# podemos o tamanho de uma lista:
tamanho = len(lista)
print(tamanho)

# percorrendo a lista por índices:
for indice in range(tamanho):
    print('índice =', indice, '| elemento =', lista[indice])
    
    
# segunda forma
print('------------------')

for elemento in lista:
    print(elemento)
```

```python
# Faça um programa que pergunta quantas provas foram realizadas, pede as notas, calcula sua média e mostra a maior e menor nota

# criando uma lista vazia
notas = []

# perguntando a quantidade:
quantidade = int(input('Quantas provas foram aplicadas: '))

# lendo as notas pelo teclado:
for contador in range(quantidade): # um loop que apenas executa "quantidade" vezes - NÃO estamos percorrendo a lista!
    nota = float(input('Digite a nota: '))
    notas.append(nota) # joga a nota lida para a lista
    
# calculando a média:
soma = 0
for n in notas: # agora sim é um loop percorrendo a lista!
    soma += n # abreviação de soma = soma + n
media = soma/quantidade # soma/len(notas) também daria certo :)


# pegando maior/menor valor de uma lista:
maior = max(notas)
menor = min(notas)

# mostrando os resultados:
print('Notas:', notas)
print('Maior nota:', maior)
print('Menor nota:', menor)
print('Média:', media)
```

```python
'''
Faça um programa que, dadas duas listas de mesmo tamanho, 
crie uma nova lista com cada elemento igual a soma dos elementos da lista 1 com os da lista 2, na mesma posição.

Exemplo:

Dadas lista1 = [1, 4, 5] e lista2 = [2, 2, 3], então lista3 = [1+2, 4+2, 5+3] = [3, 6, 8]
'''

lista1 = [1, 4, 5]
lista2 = [2, 2, 3]
lista3 = []

for indice in range(len(lista1)):
    soma = lista1[indice] + lista2[indice]
    lista3.append(soma)

print(lista3)
```
