```python
# Generated by Selenium IDE
import pytest
import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities

from bs4 import BeautifulSoup

class TestTesteAula():  
  #def setup_method(self, method):
  # mudando esse método para o construtor padrão:
  def __init__(self):
    self.driver = webdriver.Chrome('c:/chromedriver.exe') # colocar o caminho do driver no seu computador!
    self.vars = {}
  
  def teardown_method(self, method):
    self.driver.quit()
  
  def test_testeAula(self):
    self.driver.get("https://www.google.com.br/")
    self.driver.set_window_size(1340, 629)
    self.driver.find_element(By.NAME, "q").click()
    self.driver.find_element(By.NAME, "q").send_keys("Let's Code python")
    self.driver.find_element(By.NAME, "q").send_keys(Keys.ENTER)
    self.driver.execute_script("window.scrollTo(0,300)")
    self.driver.execute_script("window.scrollTo(0,600)")
    self.driver.execute_script("window.scrollTo(0,1000)")
    
objetoTeste = TestTesteAula()
objetoTeste.test_testeAula()

html = objetoTeste.driver.page_source # pegando o HTML *depois* de já ter interagido com o site

sopa = BeautifulSoup(html, 'html.parser')

descricoes = sopa.find_all('div', class_='VwiC3b')

for descricao in descricoes:
    print(descricao.text)
```
