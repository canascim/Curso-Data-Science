```python
import numpy as np
import pandas  as pd
import seaborn as sns
import scipy.stats as sts
import random
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 252)
S0 = 100
mu = 10/100
sigma = 18/100
T = 1
n_steps = 252

dt = T/n_steps




mu1, sigma1 = 0, 1 # mean and standard deviation
W = np.random.normal(mu1, sigma1, 252)
W
```

```python
import math

St_GBM = [S0]


for i in range(0,251):
    SG = S0 * np.exp((mu - (sigma ** 2)/2) * (dt) + sigma * W[i] * math.sqrt(dt))
    St_GBM.append(SG)
    
plt.figure(figsize = (20,12))
plt.plot(x, St_GBM)
plt.show()
```

```python
#Obtendo os dados para equao aritmtica


St_ABM = [S0]


for i in range(0,251):
    SA = S0 + (mu * dt) + (sigma * W[i] * math.sqrt(dt))
    St_ABM.append(SA)
    
plt.figure(figsize = (20,12))
plt.plot(x, St_ABM)
plt.show()
```

```python
#Gerando 10 mil simulaes 

numT = 10000

W_sim = []  #Lista para armazenar os arrays das simulaes dos choques

for i in range(numT):    
    W = np.random.normal(mu1, sigma1, 252)
    W_sim.append(W)


St_GBM_sim = [] #Lista com as 10 mil trajetrias de preos que sero plotadas

for Wi in W_sim:
    St_GBM = S0 * np.exp((mu - (sigma ** 2)/2) * (dt) + sigma * Wi * math.sqrt(dt))
    St_GBM_sim.append(St_GBM)


plt.figure(figsize=(20,12))
plt.subplot()
plt.plot(St_GBM_sim)
plt.show()
```

```python
#Pegando o ltimo elemento de cada uma das 10000 simulaes

St_GBM_sim_final = []

for j in range(0,10000):
    St_GBM_sim_final.append(St_GBM_sim[-j][-1])


#Agora, esses valores simulados sero trazidos a valor presente usando-se
#mu como a taxa livre de risco

valor_presente = []

for i in St_GBM_sim_final:
    valor_presente.append(i*np.exp(-mu*T))
    
valor_presente
```
